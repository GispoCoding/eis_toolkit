```python
import sys
sys.path.insert(0, "..")
from eis_toolkit.vector_processing.extract_shared_lines import extract_shared_lines

from shapely.geometry import Polygon
import geopandas as gpd

import matplotlib.pyplot as plt
```

```python
# Create polygons, add to gdf, call extract_shared_lines and plot results.

poly1 = Polygon([(0, 0), (0, 1), (1, 1), (1, 0)])
poly2 = Polygon([(1, 0), (1, 1), (2, 1), (2, 0)])
poly3 = Polygon([(2, 0), (2, 1), (3, 1), (3, 0)])

data = {'geometry': [poly1, poly2, poly3]}
gdf = gpd.GeoDataFrame(data)

extracted_shared_lines_gdf = extract_shared_lines(gdf)

fig, ax = plt.subplots()

gdf.plot(ax=ax, color='lightblue', edgecolor='black', alpha=0.5)

if not extracted_shared_lines_gdf.empty:
    extracted_shared_lines_gdf.plot(ax=ax, color='red')

ax.set_xlim([-1, 4])
ax.set_ylim([-1, 2])

plt.show()
```
